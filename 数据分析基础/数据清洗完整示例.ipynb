```python
import pandas as pd
import numpy as np

# 1. 患者基本信息表（含重复值、缺失值）
patients = pd.DataFrame({
    '患者ID': [1, 2, 3, 4, 5, 6, 7, 8],
    '姓名': ['张三', '李四', '王五', '赵六', '孙七', '周八', '吴九', '郑十'],
    '年龄': [45, 67, 52, 38, 29, 55, None, 61],
    '性别': ['男', '女', '男', '女', '男', '女', '男', '女'],
    '身高': [175, 160, 172, 165, 180, 158, 170, None],
    '体重': [80, 65, 78, 55, 85, 52, 72, 68]
})

# 2. 就诊记录表（含重复行、缺失值）
visits = pd.DataFrame({
    '患者ID': [1, 1, 2, 2, 3, 3, 4, 5, 5, 6, 7, 8, 8],
    '就诊日期': ['2025-01-01', '2025-06-01', '2025-01-15', '2025-07-01',
                 '2025-02-01', '2025-06-15', '2025-01-20',
                 '2025-03-01', '2025-08-01', '2025-04-01',
                 '2025-05-01', '2025-02-15', '2025-02-15'],  # 故意重复一行
    '收缩压': [130, 125, 140, 135, 120, 118, 150, 125, 130, 140, 135, 128, None],
    '舒张压': [85, 82, 90, 88, 80, 78, 95, 82, 85, 88, 85, 82, 80],
    '血糖': [5.6, 5.2, 7.8, 7.0, 4.5, 6.0, 6.5, 5.0, 4.8, 6.2, 5.5, 5.0, 5.1]
})

print("===== 原始患者表 =====")
print(patients)
print("\n===== 原始就诊表 =====")
print(visits)
```

```python
# 1. 查看前几行
print("患者表前3行：")
print(patients.head(3))
print("\n就诊表前3行：")
print(visits.head(3))

# 2. 查看整体信息（列名、类型、非空数量）
print("\n患者表信息：")
print(patients.info())
print("\n就诊表信息：")
print(visits.info())

# 3. 查看缺失值统计
print("\n患者表缺失值：")
print(patients.isna().sum())
print("\n就诊表缺失值：")
print(visits.isna().sum())
```

```python
# 1. 处理患者表缺失值
print("处理前患者表缺失值：")
print(patients.isna().sum())

# 年龄缺失 → 填平均值（整数）
patients['年龄'] = patients['年龄'].fillna(patients['年龄'].mean()).astype(int)

# 身高缺失 → 填平均值（保留1位小数）
patients['身高'] = patients['身高'].fillna(patients['身高'].mean()).round(1)

print("\n处理后患者表缺失值：")
print(patients.isna().sum())

# 2. 处理就诊表缺失值
# 收缩压缺失 → 填中位数
visits['收缩压'] = visits['收缩压'].fillna(visits['收缩压'].median())

print("\n处理后就诊表缺失值：")
print(visits.isna().sum())

# 3. 删除就诊表中的完全重复行
visits = visits.drop_duplicates()
print(f"\n删除重复行后就诊表行数：{len(visits)}")
```

```python
# 1. 就诊日期转为日期类型
visits['就诊日期'] = pd.to_datetime(visits['就诊日期'])

# 2. 按患者ID和就诊日期排序（便于后续取最新记录）
visits = visits.sort_values(['患者ID', '就诊日期'])
print("\n排序后的就诊表（按患者ID + 日期）：")
print(visits)
```

```python
# 按患者ID统计：总就诊次数、平均收缩压、平均血糖
visits_stats = visits.groupby('患者ID').agg({
    '收缩压': ['count', 'mean'],
    '血糖': 'mean'
})
# 重命名列名
visits_stats.columns = ['就诊次数', '平均收缩压', '平均血糖']
visits_stats = visits_stats.reset_index()

print("\n===== 分组统计结果 =====")
print(visits_stats)
```

```python
# 先给每次就诊编号（按日期排序后，每个患者从1开始编号）
visits['就诊序号'] = visits.groupby('患者ID').cumcount() + 1

# 用 pivot_table 取出最后一次就诊（序号最大的那次）
latest_wide = visits[visits['就诊序号'] == visits.groupby('患者ID')['就诊序号'].transform('max')]

# 转成宽格式（每个患者一行，指标作为列）
latest_pivot = latest_wide.pivot_table(
    index='患者ID',
    values=['收缩压', '舒张压', '血糖'],
    aggfunc='first'
).reset_index()

# 拍平列名
latest_pivot.columns = [f'最新_{col}' if col != '患者ID' else col for col in latest_pivot.columns]

print("\n===== 最新就诊指标（宽格式） =====")
print(latest_pivot)
```

```python
# 1. 先合并患者表和分组统计表
merged = pd.merge(patients, visits_stats, on='患者ID', how='left')

# 2. 再合并最新就诊指标
merged = pd.merge(merged, latest_pivot, on='患者ID', how='left')

print("\n===== 合并后的完整表 =====")
print(merged)
```

```python
# 1. 计算 BMI = 体重 / (身高/100)^2
def calculate_bmi(row):
    if pd.notna(row['身高']) and row['身高'] > 0:
        return round(row['体重'] / ((row['身高'] / 100) ** 2), 1)
    return None

merged['BMI'] = merged.apply(calculate_bmi, axis=1)

# 2. 年龄分组
def age_group(age):
    if age < 40:
        return '青年'
    elif age < 60:
        return '中年'
    else:
        return '老年'

merged['年龄分组'] = merged['年龄'].apply(age_group)

print("\n===== 最终特征表（含BMI和年龄分组） =====")
print(merged)
```

```python
# 保存清洗后的数据到 CSV
merged.to_csv('cleaned_patient_data.csv', index=False, encoding='utf-8-sig')
print("\n清洗后的数据已保存为 cleaned_patient_data.csv")
```
