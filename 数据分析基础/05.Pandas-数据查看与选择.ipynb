```python
import pandas as pd

df=pd.read_csv("data/sales.csv",usecols=["订单号","产品类别","产品名称","销售数量","单价"])

df.head()#查看前几行数据，默认是5
df.tail(8)#查看后几行数据，默认是5
df.describe()#查看df中的各种数值的统计信息
df.info()#查看df中各个列名，各个列中元素是否非空，以及数据类型
df.shape#查看df中有几列几行（列，行）
df.columns#显示列名
```

```python
#数据选择
df2=pd.read_csv("data/sales.csv",usecols=["订单号","产品类别","产品名称","销售数量","单价"],index_col=["订单号"])
df2

# 选择列
# 1.单列
# df2["销售数量"]
# 2.多列
df2[["单价","销售数量"]]

#选择行
#1.根据行号选择。语法：df.iloc[start:stop:step]------>step默认为1，输出不包括第stop行
df2.iloc[0:5]
#2.根据索引选择。语法：df.loc[start:stop:step]------>step默认为1，输出包含stop行
df2.loc[6805677496:6898807429:1]
```
